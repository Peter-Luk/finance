```python
from bokeh.plotting import figure, output_notebook, show
code, cmatch, omatch = 'mhik7', {'SMA':'blue', 'WMA':'green', 'EMA':'cyan', 'KAMA':'purple'}, {'KC':'purple', 'APZ':'cyan', 'BB':'blue'}
mp = getattr(__import__('pt_2'),'PI')(code, 12)
imp, omp = mp.fdc('i'), mp.fdc('o')
output_notebook()
```

```python
q = figure(title='%s daily' % code.upper(), x_axis_label='Date', plot_width=600, plot_height=250, x_axis_type='datetime')
q.line(imp['Date'], imp['RSI'], legend='RSI', color='navy', line_width=2, alpha=0.5)
q.legend.location = 'top_left'
show(q)
```

```python
bmp = mp.fdc('b')
inc = bmp.Close > bmp.Open
dec = bmp.Open > bmp.Close
w = 12 * 60 * 60 * 1000
TOOLS = 'pan,wheel_zoom,box_zoom,reset,save'
r = figure(x_axis_type='datetime', tools=TOOLS, plot_width=600, title='%s Candlestick with Overlays' % code.upper(), responsive=True)
r.grid.grid_line_alpha = 0.3
r.segment(bmp.Date, bmp.High, bmp.Date, bmp.Low, color='black')
r.vbar(bmp.Date[inc], w, bmp.Open[inc], bmp.Close[inc], fill_color='blue', line_color='black')
r.vbar(bmp.Date[dec], w, bmp.Open[dec], bmp.Close[dec], fill_color='red', line_color='black')
for k, v in omatch.items():
    l, h = [], []
    if k == 'KC':
        for i in omp.KC.values:
            l.append(i[0])
            h.append(i[-1])
    if k == 'BB':
        for i in omp.BB.values:
            l.append(i[0])
            h.append(i[-1])
    if k == 'APZ':
        for i in omp.APZ.values:
            l.append(i[0])
            h.append(i[-1])
    r.line(omp['Date'], l, color=v, line_width=2, alpha=0.5)
    r.line(omp['Date'], h, color=v, legend=k, line_width=2, alpha=0.5)
r.legend.location = 'top_right'
show(r)
```

```python
p = figure(x_axis_type='datetime', tools=TOOLS, plot_width=600, title='%s Candlestick with MAs' % code.upper(), responsive=True)
p.grid.grid_line_alpha = 0.3
p.segment(bmp.Date, bmp.High, bmp.Date, bmp.Low, color='black')
p.vbar(bmp.Date[inc], w, bmp.Open[inc], bmp.Close[inc], fill_color='blue', line_color='black')
p.vbar(bmp.Date[dec], w, bmp.Open[dec], bmp.Close[dec], fill_color='red', line_color='black')
[p.line(imp['Date'], imp[k], legend=k, color=v, line_width=2, alpha=0.5) for k, v in cmatch.items()]
p.legend.location = 'top_left'
show(p)
```
